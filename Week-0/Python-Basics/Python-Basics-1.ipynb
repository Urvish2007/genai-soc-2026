# Function applied to function means we are passing function as argument into another function. 

# function that operators on other functions are called "higher-order-functions."

```python
def mult_by_five(x):
    return 5 * x

def call(fn, arg):
    """Call fn on arg"""
    return fn(arg)

def squared_call(fn, arg):
    """Call fn on the result of calling fn on arg"""
    return fn(fn(arg))

print(
    call(mult_by_five, 1),
    squared_call(mult_by_five, 1), 
    sep='\n', # '\n' is the newline character - it starts a new line
)
```

# as we know builtin max function returns maximum among all numbers that have passed as an argument but now if you use key (keyword) it would going to give you number among them which maximise that function
